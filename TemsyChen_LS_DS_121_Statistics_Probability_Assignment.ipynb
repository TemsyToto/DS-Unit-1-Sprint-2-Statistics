### 2) Pretend that this dataset represents the cars at a used car lot in your local town. 

- Use df.sample() to pick a random sample of 10 cars. Note that because this sample is random we are going to set the `random_state` so that all of us in the class get the same random sample. Please set your random state to `30` when using `df.sample()`

Is your sample reflective of the population value in regards to highway-mpg? Find the mean for `highway-mpg` for the entire dataset and use a 1-sample t-test to compare your estaimated sample mean to the population mean. Can you say that they are different? 

```python
### YOUR WORK HERE
sample = df.sample(n=10, random_state=30)
sample
```

```python
#find the mean of the highway mpg

sample_mean = sample['highway mpg'].mean()
sample_mean
```

```python
sample.info()
```

```python
#The mean of the population highway mpg

df_mean = df['highway mpg'].mean()
df_mean
```

```python
#Now do the t-test

from scipy import stats

stats.ttest_1samp(a= sample['highway mpg'], popmean= df_mean)

#Answer: p-value is .4, so I fail to reject the null, which means the sample highway mpg is reflective of the population highway mpg
```

The salesman says the cars he sells typically have a fuel efficiency of about 35 miles per gallon on the highway. You want to verify his claim but can only test 10 cars. Using your sample of 10, test his claim and report your results.

```python
### YOUR WORK HERE

stats.ttest_1samp(a= sample['highway mpg'], popmean= 35)

#p-value is .36, I fail to reject the null, which means his cars typically have a fuel effiency of 35 mpg, like the general population.
```

Imagine that you now have the capacity to test 100 cars on the lot. Using the same random state of `30`, take a sample of 100 cars. Run a t-test to verify the salesman's claim with your new larger sample. Do you reach the same conclusion as you did with the sample size of 10? 

```python
### YOUR WORK HERE

sample_100 = df.sample(n=100, random_state=30)
sample_100
```

```python
#t-test on the salesman's claim against a larger sample mean

stats.ttest_1samp(a= sample_100['highway mpg'], popmean= 35)
```

## Practice 2-Sample T-tests

Two Sample t-tests determine whether or not two sample means are statistically different from each other. 

This portion of your assignment requires you to determine which issues have "statistically significant" differences between political parties in this [1980s congressional voting data](https://archive.ics.uci.edu/ml/datasets/Congressional+Voting+Records). 


### 1) Load the data

The data consists of 435 instances (one for each congressperson), a class (democrat or republican), and 16 binary attributes (yes or no for voting for or against certain issues). Be aware - there are missing values!

- Read the dataset in from UCI, you'll need to provide a list of column headers
- Encode "yes" votes as 1 and "no" votes as 0. (You can use `df.replace()` to swap out these values)
- Use dataframe filtering to split the dataframe into two new dataframes based on party. Hold all republicans in one dataframe and all democrats in the other. These will be our two different "samples."



```python
### YOUR WORK HERE
#read the file, noticed the column header was a data row, so entered header as none

votes = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/voting-records/house-votes-84.data', header= None)
votes
```

```python
#rename the columns

votes = votes.rename(columns={0:'class name',1:'handicapped infants',2:'water project cost sharing',3:'adoption of budget res',4:'physician fee freeze',
                              5:'el salvador aid',6:'religious groups in schools',7:'anti satellite test ban',8:'aid to nicaraguan contras',9:'mx missile',
                              10:'immigration',11:'synfuels corporation cutback',12:'education spending',13:'superfund right to sue',14:'crime',15:'duty free exports',
                              16:'export admin act south africa'})
votes
```

```python
#replace the ? with NaN, n with 0, and y with 1

votes_clean = votes.replace("?",  np.NaN).replace('n', 0).replace('y', 1)
votes_clean
```

```python
#create two new dataframes, one republican, one democrat

condition1 = (votes_clean['class name'] == 'republican')
republican = votes_clean[condition1]

condition2 = (votes_clean['class name'] == 'democrat')
democrat = votes_clean[condition2]
```


### 2) Perform two sample T-tests on different issues and report the results.

- Find an issue that democrats support more than republicans with p < 0.01 (significant at the 99% level).
- Find an issue that republicans support more than democrats with p < 0.01 (significant at the 99% level).
- Find an issue where the difference between republicans and democrats has p > 0.1 (Not significant at the 90% level - i.e. there may not be much of a difference the two sample means)

Please for each test that you run state your null and alternative hypothesis and then write a conclusion reflecting on the null and alternative hypothesis.

Remember, that two-sample t-tests will only tell us if the two groups are *different* from one another. We'll have to look at their sample means directly or use the sign on the t-statistic to know which of the two sample means is larger. 

```python
### YOUR WORK HERE

#Null hypothesis: "handicapped infants" is an issue that democrats vote the same as republicans
#Alternative hypothesis: this is an issue where dems vote differently than reps.

print("2 sample t-test:", stats.ttest_ind(democrat['handicapped infants'], republican['handicapped infants'], nan_policy='omit'))
print("Dem vote mean:", democrat['handicapped infants'].mean())
print("Rep vote mean:", republican['handicapped infants'].mean())

#Result: Dems supported this issue more than Reps, with a p-value less than 0.01
#So I reject the null hypothesis, they do not vote the same on this issue.
```

```python
#Null hypothesis: "religious groups in schools" is an issue that democrats vote the same as republicans
#Alternative hypothesis: this is an issue where dems vote differently than reps.
print("2 sample t-test:", stats.ttest_ind(democrat['religious groups in schools'], republican['religious groups in schools'], nan_policy='omit'))
print("Dem vote mean:", democrat['religious groups in schools'].mean())
print("Rep vote mean:", republican['religious groups in schools'].mean())

#Result: Reps supported this issue more than Dems, with a p-value less than 0.01
```

```python
#Null hypothesis: "water project cost sharing" is an issue that democrats vote the same as republicans
#Alternative hypothesis: this is an issue where dems vote differently than reps.

print("2 sample t-test:", stats.ttest_ind(democrat['water project cost sharing'], republican['water project cost sharing'], nan_policy='omit'))
print("Dem vote mean:", democrat['water project cost sharing'].mean())
print("Rep vote mean:", republican['water project cost sharing'].mean())

#Result: Reps and Dems vote so similarly on this, there is a p-value .93, so I fail to reject the null.
```

```python
#A way to find the mean of every column quickly

print("Dem Votes:\n", democrat.mean(),"\n")
print("Rep Votes:\n", republican.mean())
```
